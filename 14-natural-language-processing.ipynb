```python
# tahapan NLP
# 1. string data
# 2. cleansing (manipulasi kata dari string)
# 3. stemming (konversi kata menjadi kata dasar)
# 4. tokenizing (merubah ke array)
# 5. stopwords
# 6. output

import requests
import string
import re

from bs4 import BeautifulSoup
import nltk
from nltk.corpus import stopwords

!pip install Sastrawi
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
```

```python
# web = requests.get('https://tirto.id/rumah-bangsa-viking-gSHt').text
web = requests.get('https://tirto.id/').text
soup = BeautifulSoup(web)

for s in soup(['script', 'style']):
  s.decompose()

text = ' '.join(soup.stripped_strings)
print(text)
```

```python
# cleaning
text = text.lower()
text = re.sub(r"\d+", "", text) # cleaning from number
text = text.translate(str.maketrans("", "", string.punctuation)) # cleaning from punctuation
text = text.strip() # cleaning from empty character
print("After cleanning:\n", text, "\n")

# stemming
factory = StemmerFactory()
stemmer = factory.create_stemmer()
output = stemmer.stem(text)
print("After stemming:\n", output)
```

```python
# tokenizing
tokens = [t for t in output.split()]
print("After tokenizing:\n", tokens, "\n")
```

```python
nltk.download()
clean_tokens = tokens[:]

for token in tokens:
  if token in stopwords.words('indonesian'):
    clean_tokens.remove(token)

freq = nltk.FreqDist(clean_tokens)

for key, val in freq.items():
  print(str(key), ":", str(val))

freq.plot(30)
```
