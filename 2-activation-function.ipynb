```python
# table logika AND 2 input
"""

x1    x2    y
1     1     1
1     0     0
0     1     0
0     0     0

"""
```

```python
# machine learning

import keras
import numpy as np
import matplotlib.pyplot as plt

# model
model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[2])])

# compile
model.compile(optimizer='sgd', loss='mean_squared_error')
```

```python
class LossHistory(keras.callbacks.Callback):
  def on_train_begin(self, logs={}):
    self.losses = []

  def on_batch_end(self, batch, logs={}):
    self.losses.append(logs.get('loss'))
```

```python
# set data
xs = np.array([[1,1], [1,0], [0,1], [0,0]], dtype=int)
ys = np.array([1,0,0,0], dtype=int)
```

```python
# tampilkan arsitektur awal
model.summary()
```

```python
# training
history = LossHistory()
model.fit(xs, ys, epochs=100, callbacks=[history])
```

```python
plot1 = plt.figure(1)
plt.plot(history.losses)
plt.show()
```

```python
# model 2
model2 = keras.Sequential([keras.layers.Dense(units=1, input_shape=[2], activation=keras.activations.tanh)])

# compile
model2.compile(optimizer='sgd', loss='mean_squared_error')
```

```python
# training
history2 = LossHistory()
model2.fit(xs, ys, epochs=100, callbacks=[history2])
```

```python
plot2 = plt.figure(2)
plt.plot(history2.losses)
plt.show()
```

```python
# comparation before and after ativation

plot3 = plt.figure(3)

plt.xlabel('epoch')
plt.ylabel('loss')

plt.plot(history.losses, label='linier')
plt.plot(history2.losses, label='tanh')
plt.legend()
plt.show()
```
